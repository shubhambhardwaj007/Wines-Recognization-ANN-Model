```python
#Feature Elimination
x=dataset
for i in ["Ash","Magnesium","Nonflavanoid phenols","Classes"]:
    x=x.drop(i,axis=1)
x
```

```python
x.shape
```

```python
x.info()
```

```python
from keras.models import Sequential
```

```python
model = Sequential()
```

```python
model.get_config()
```

```python
from keras.layers import Dense
```

```python
model.add(
    Dense(
        units = 16,
        input_shape=(10,),
        activation="relu",
        kernel_initializer="he_normal"
    )
)
```

```python
model.summary()
```

```python
model.add(
    Dense(
        units=16,
        activation='relu',
        kernel_initializer='he_normal'
    )
)
```

```python
model.summary()
```

```python
model.add(
    Dense(
        units=16,
        activation='relu',
        kernel_initializer='he_normal'
    )
)
```

```python
model.get_config()
```

```python
model.add(
    Dense(
        units=3,
        activation="softmax"
    )
)
```

```python
model.summary()
```

```python
from keras.optimizers import RMSprop
```

```python
model.compile(optimizer=RMSprop(),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x,y_categorical,epochs=275)
```

```python
model.history.history
```

```python
loss=model.history.history['loss']
```
